```python
from rich.tree import Tree
from rich.table import Table
from plotly import graph_objects as go
import json

def rich_table_to_html(table):
    html = ['<table>']

    # Add the header row.
    html.append('<thead><tr>')
    for column in table.columns:
        html.append(f'<th>{column.header}</th>')
    html.append('</tr></thead>')

    # Add the data rows.
    html.append('<tbody>')
    for row in table.rows:
        html.append('<tr>')
        for cell in row.cells:
            html.append(f'<td>{cell}</td>')
        html.append('</tr>')
    html.append('</tbody>')

    html.append('</table>')
    return ''.join(html)

def process_node(node):
    name = node.label
    children = node.children
    if children:
        return {
            "name": name,
            "children": [process_node(child) for child in children],
            "table": rich_table_to_html(node) if isinstance(node, Table) else None
        }
    else:
        return {
            "name": name,
            "table": rich_table_to_html(node) if isinstance(node, Table) else None
        }


def rich_tree_to_dendrogram(tree):
    processed_tree = process_node(tree)
    # Turn it into a json string for ease of use in JavaScript
    json_tree = json.dumps(processed_tree)

    dendrogram_html = f"""
    <div id="dendrogram"></div>
    <script>
        var treeData = {json_tree};

        var layout = {{
            margin: {{l: 0, r: 0, b: 0, t: 0}},
            sunburstcolorway: ["#636efa", "#ef553b", "#00cc96"],
        }};

        var data = [treeData];

        var graph = {{
            data: data,
            layout: layout
        }};

        Plotly.newPlot('dendrogram', graph.data, graph.layout);
    </script>
    """
    return dendrogram_html


from IPython.display import display, HTML

# tree is your Rich Tree object
tree = towers["user"].__rich_repr__()
# dendrogram_html = rich_tree_to_dendrogram(tree)


process_node(tree)
# display(HTML(dendrogram_html))

# tree
```

```python
from IPython.display import display
from ipywidgets import widgets
from rich.tree import Tree
from rich.table import Table

# Helper function to convert a Rich Table to HTML
def rich_table_to_html(table):
    html = ['<table>']

    # Add the header row.
    html.append('<thead><tr>')
    for column in table.columns:
        html.append(f'<th>{column.header}</th>')
    html.append('</tr></thead>')

    # Add the data rows.
    html.append('<tbody>')
    for row in table.rows:
        html.append('<tr>')
        for cell in row.cells:
            html.append(f'<td>{cell}</td>')
        html.append('</tr>')
    html.append('</tbody>')

    html.append('</table>')
    return ''.join(html)

def process_node(node):
    name = node.label
    children = node.children
    items = []
    if isinstance(node, Table):
        items.append(widgets.HTML(rich_table_to_html(node)))
    for child in children:
        items.append(process_node(child))
    return widgets.VBox([widgets.HTML(f"<b>{name}</b>")] + items)

def rich_tree_to_widget(tree):
    return process_node(tree)

# tree is your Rich Tree object
tree_widget = rich_tree_to_widget(tree)
# display(tree_widget)
tree_widget
```
